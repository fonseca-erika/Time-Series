Time Series - Neural Network Auto-Regression
</div></h1>

<b>Non seasonal data:</b> $NNAR_{p,k}$ model:
- Inputs: lagged values of the time series $x_t , . . . , x_{t−p}$
- 1 hidden layer with k neurons
Rk: $NNAR_{p,0} = AR_p$

<b>Seasonal data</b> (of period T): $NNAR_{(p,P,k)T}$ model
we add lagged values from the same season as last observed values:
- Inputs: lagged values of the time series $x_t , x_{t−2}, . . . , x_{t−p}, x{t−T} , x_{t−2T} , . . . , x_{t−pT}$
- 1 hidden layer with k neurons
Rk: $NNAR_{(p,P,0)T} = SARIMA_{(p,0,0)(P,0,0)T}$
